```python
import zipfile
import os
from tqdm import tqdm

# 1. ì„¤ì • (ì•„ê¹Œì™€ ë™ì¼)
zip_file_path = "cifake-real-and-ai-generated-synthetic-images.zip"
extract_path = dataset_path  # êµ¬ê¸€ ë“œë¼ì´ë¸Œ ê²½ë¡œ

print(f"ğŸš‘ ë°ì´í„° ì‘ê¸‰ ë³µêµ¬ ì‹œì‘: {extract_path}")

# 2. ëˆ„ë½ëœ íŒŒì¼ ì°¾ê¸° ë° ë³µêµ¬
try:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        # Zip ì•ˆì— ìˆëŠ” ëª¨ë“  íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
        all_files = zip_ref.namelist()

        # 'train/REAL' í´ë”ì— í¬í•¨ëœ íŒŒì¼ë§Œ ê³¨ë¼ë‚´ê¸° (ì´ë¯¸ì§€ íŒŒì¼ë§Œ)
        # CIFAKE ë°ì´í„°ì…‹ì€ ë³´í†µ 'train/REAL' ë˜ëŠ” 'REAL' ì´ë¦„ì´ ê²½ë¡œì— ë“¤ì–´ê°‘ë‹ˆë‹¤.
        missing_targets = [f for f in all_files if "REAL" in f and (f.endswith('.jpg') or f.endswith('.png'))]

        print(f"ğŸ§ Zip íŒŒì¼ ë‚´ REAL ì´ë¯¸ì§€ ë°œê²¬: {len(missing_targets):,}ê°œ")

        if len(missing_targets) == 0:
            print("âš ï¸ Zip íŒŒì¼ ì•ˆì—ë„ REAL í´ë”ê°€ ì•ˆ ë³´ì…ë‹ˆë‹¤! ê²½ë¡œëª…ì„ ë‹¤ì‹œ í™•ì¸í•´ì•¼ í•©ë‹ˆë‹¤.")
            # í˜¹ì‹œ í´ë”ëª…ì´ ì†Œë¬¸ì real ì¼ìˆ˜ë„ ìˆìœ¼ë‹ˆ í™•ì¸
            print("   (ì°¸ê³ ) Zip ë‚´ë¶€ ìµœìƒìœ„ í´ë”ë“¤:", set([f.split('/')[0] for f in all_files[:20]]))
        else:
            print("ğŸš€ ëˆ„ë½ëœ REAL ì´ë¯¸ì§€ ì¶”ì¶œ ì‹œì‘...")

            count = 0
            for file in tqdm(missing_targets, desc="Restoring REAL", unit="img"):
                # íŒŒì¼ì´ ì´ë¯¸ ë“œë¼ì´ë¸Œì— ìˆëŠ”ì§€ í™•ì¸ (ìˆìœ¼ë©´ ìŠ¤í‚µí•´ì„œ ì‹œê°„ ì ˆì•½)
                target_full_path = os.path.join(extract_path, file)

                if not os.path.exists(target_full_path):
                    zip_ref.extract(file, extract_path)
                    count += 1

            print(f"\nâœ… ë³µêµ¬ ì™„ë£Œ! ì´ {count}ì¥ì˜ REAL ì´ë¯¸ì§€ë¥¼ ì¶”ê°€ë¡œ ì €ì¥í–ˆìŠµë‹ˆë‹¤.")

except Exception as e:
    print(f"\nâŒ ì˜¤ë¥˜ ë°œìƒ: {e}")

# 3. ìµœì¢… ê°œìˆ˜ ì¬í™•ì¸ (ê²€ì¦)
print("\nğŸ“Š [ìµœì¢… ê²€ì¦] ë°ì´í„°ì…‹ ê°œìˆ˜ í™•ì¸")
for split in ['train', 'test']:
    for label in ['REAL', 'FAKE']:
        folder_path = os.path.join(extract_path, split, label)
        if os.path.exists(folder_path):
            print(f"ğŸ‘‰ {split}/{label}: {len(os.listdir(folder_path)):,} ì¥")
        else:
            print(f"âŒ {split}/{label}: ì—¬ì „íˆ í´ë”ê°€ ì—†ìŠµë‹ˆë‹¤!")
```

```python
import os

# ì•„ê¹Œ ì„¤ì •í•œ ê²½ë¡œ (í˜¹ì‹œ ë³€ìˆ˜ê°€ ì§€ì›Œì¡Œì„ê¹Œë´ ë‹¤ì‹œ ì§€ì •)
# ë³¸ì¸ì˜ ê²½ë¡œì— ë§ê²Œ ìˆ˜ì •ì´ í•„ìš”í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# ë³´í†µ '/content/drive/MyDrive/Hecto_Deepfake' ë˜ëŠ” 'dataset_root' ì…ë‹ˆë‹¤.
target_path = dataset_path

print(f"ğŸ“‚ ë°ì´í„°ì…‹ ê²½ë¡œ í™•ì¸: {target_path}\n")
print("ğŸ“Š í´ë”ë³„ ì´ë¯¸ì§€ ê°œìˆ˜ ì§‘ê³„ ì¤‘...")
print("-" * 30)

total_train = 0
total_test = 0

# trainê³¼ test í´ë” ê°ê° í™•ì¸
for split in ['train', 'test']:
    split_total = 0
    for label in ['REAL', 'FAKE']:
        # ê²½ë¡œ: .../train/REAL, .../train/FAKE í˜•íƒœ
        folder_check = os.path.join(target_path, split, label)

        if os.path.exists(folder_check):
            # íŒŒì¼ ê°œìˆ˜ ì„¸ê¸°
            count = len([f for f in os.listdir(folder_check) if os.path.isfile(os.path.join(folder_check, f))])
            print(f"âœ… [{split.upper()}] - {label}: {count:,} ì¥")
            split_total += count
        else:
            print(f"âŒ [{split.upper()}] - {label}: í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤! ({folder_check})")

    print(f"   ğŸ‘‰ {split.upper()} ì „ì²´ í•©ê³„: {split_total:,} ì¥")
    if split == 'train': total_train = split_total
    else: total_test = split_total
    print("-" * 30)

print(f"ğŸ“ˆ ì´ ë°ì´í„° ê°œìˆ˜: {total_train + total_test:,} ì¥")
```

```python
import cv2
import os
import torch
from facenet_pytorch import MTCNN
from tqdm import tqdm
from PIL import Image
import glob
import numpy as np

# ==========================================
# 1. ì„¤ì • (Configuration)
# ==========================================
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"ğŸš€ ì‘ì—… ì¥ì¹˜: {device} (GPUê°€ ì¼œì ¸ ìˆì–´ì•¼ ë¹ ë¦…ë‹ˆë‹¤!)")

# MTCNN (ì–¼êµ´ íƒì§€ê¸°) ë¡œë“œ
# margin=0: ì–¼êµ´ì„ íƒ€ì´íŠ¸í•˜ê²Œ ìë¦…ë‹ˆë‹¤. (ë°°ê²½ ìµœì†Œí™”)
mtcnn = MTCNN(keep_all=False, select_largest=True, margin=10, post_process=False, device=device)

# ê²½ë¡œ ì„¤ì •
base_path = '/content/drive/MyDrive/DILAB/OK/hecto-deepfake-detection/dataset'
source_root = os.path.join(base_path, 'dataset_celeb')   # ë™ì˜ìƒ ì°¾ëŠ” ê³³
dest_root = os.path.join(source_root, 'cropped_faces')    # ì €ì¥í•  ê³³

print(f"ğŸ“‚ ë™ì˜ìƒ ê²€ìƒ‰ ê²½ë¡œ: {source_root}")
print(f"ğŸ“‚ ì´ë¯¸ì§€ ì €ì¥ ê²½ë¡œ: {dest_root}")

# ==========================================
# 2. ì–¼êµ´ ì¶”ì¶œ í•¨ìˆ˜ (í•µì‹¬ ë¡œì§)
# ==========================================
def extract_faces_from_videos():
    # 1. ëª¨ë“  mp4 íŒŒì¼ ì°¾ê¸° (ì¬ê·€ ê²€ìƒ‰)
    video_files = glob.glob(os.path.join(source_root, '**', '*.mp4'), recursive=True)

    if not video_files:
        print("âŒ ë™ì˜ìƒ(.mp4)ì„ í•˜ë‚˜ë„ ëª» ì°¾ì•˜ìŠµë‹ˆë‹¤! ë‹¤ìš´ë¡œë“œê°€ ëœ ëê±°ë‚˜ ê²½ë¡œê°€ ë‹¤ë¦…ë‹ˆë‹¤.")
        return

    print(f"ğŸ¥ ì´ {len(video_files):,}ê°œì˜ ë™ì˜ìƒì„ ë°œê²¬í–ˆìŠµë‹ˆë‹¤. ì‘ì—…ì„ ì‹œì‘í•©ë‹ˆë‹¤...")
    print("â˜• (ì‹œê°„ì´ ê½¤ ê±¸ë¦¬ë‹ˆ ì°¨ í•œ ì” ë“œì‹œê³  ì˜¤ì„¸ìš”)")

    # 2. ë™ì˜ìƒ í•˜ë‚˜ì”© ì²˜ë¦¬
    total_extracted = 0
    error_count = 0

    for video_path in tqdm(video_files, desc="Processing Videos"):
        try:
            # ì €ì¥í•  í´ë” êµ¬ì¡° ë§Œë“¤ê¸° (ì›ë³¸ í´ë”ëª… ìœ ì§€)
            # ì˜ˆ: .../Celeb-real/id0/001.mp4 -> .../cropped_faces/Celeb-real/id0/001_frame0.jpg
            relative_path = os.path.relpath(video_path, source_root)
            # íŒŒì¼ëª… ì•ë¶€ë¶„ì´ 'raw_celeb'ì´ë©´ ê²½ë¡œì—ì„œ ì œê±°í•´ì„œ ê¹”ë”í•˜ê²Œ ë§Œë“¦
            if relative_path.startswith('raw_celeb'):
                relative_path = relative_path.replace('raw_celeb/', '')

            save_folder = os.path.dirname(os.path.join(dest_root, relative_path))
            os.makedirs(save_folder, exist_ok=True)

            # ë¹„ë””ì˜¤ íŒŒì¼ëª… (í™•ì¥ì ì œì™¸)
            video_name = os.path.splitext(os.path.basename(video_path))[0]

            # OpenCVë¡œ ë¹„ë””ì˜¤ ì—´ê¸°
            cap = cv2.VideoCapture(video_path)
            frame_count = 0

            while True:
                success, frame = cap.read()
                if not success: break

                # [ì¤‘ìš”] ëª¨ë“  í”„ë ˆì„ì„ ë‹¤ í•˜ë©´ ë„ˆë¬´ ë§ìŒ -> 30í”„ë ˆì„(ì•½ 1ì´ˆ)ë§ˆë‹¤ 1ì¥ì”©ë§Œ ì¶”ì¶œ
                if frame_count % 30 == 0:
                    # BGR -> RGB ë³€í™˜
                    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                    frame_pil = Image.fromarray(frame_rgb)

                    # ì €ì¥ íŒŒì¼ëª…
                    save_path = os.path.join(save_folder, f"{video_name}_f{frame_count}.jpg")

                    # ì´ë¯¸ ìˆìœ¼ë©´ ìŠ¤í‚µ (ì´ì–´í•˜ê¸° ê¸°ëŠ¥)
                    if not os.path.exists(save_path):
                        # ì–¼êµ´ ê°ì§€ ë° ì €ì¥ (MTCNNì´ ì•Œì•„ì„œ ë‹¤ í•´ì¤Œ)
                        mtcnn(frame_pil, save_path=save_path)
                        total_extracted += 1

                frame_count += 1
            cap.release()

        except Exception as e:
            error_count += 1
            # ì—ëŸ¬ ë‚˜ë„ ë©ˆì¶”ì§€ ì•Šê³  ë‹¤ìŒ ë™ì˜ìƒìœ¼ë¡œ ë„˜ì–´ê°
            continue

    print("\n" + "="*50)
    print(f"ğŸ‰ ì‘ì—… ì™„ë£Œ!")
    print(f"âœ… ì¶”ì¶œëœ ì–¼êµ´ ì´ë¯¸ì§€: ì•½ {total_extracted:,}ì¥ (ì¶”ì •)")
    print(f"âš ï¸ ì½ê¸° ì‹¤íŒ¨í•œ ì˜ìƒ: {error_count}ê°œ")
    print("="*50)

# ì‹¤í–‰!
extract_faces_from_videos()
```
